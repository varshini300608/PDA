```python
# HEATMAP
import seaborn as sns 
import matplotlib.pyplot as plt
import pandas as pd
data = pd.DataFrame({
'A': [1, 2, 3, 4],
'B': [4, 3, 2, 1],
'C': [2, 3, 4, 5]
})
corr = data.corr()
sns.heatmap(corr,annot=True)
plt.title( "Heatmap")
plt.show()
```
