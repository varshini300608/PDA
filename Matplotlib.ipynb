```python
#line plot 
import matplotlib.pyplot as plt


x=[1,2,3,4]
y=[10,20,30,40]

plt.plot(x,y)
plt.title("line plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
#bar chart 
import matplotlib.pyplot as plt 

categories=['a','b','c','d']
values=[5,7,3,8]

plt.bar(categories,values)
plt.title("bar chart")
plt.xlabel("categories")
plt.ylabel("values")
plt.show()
```

```python
#HISTOGRAM
import matplotlib.pyplot as plt

data=[1,2,2,3,3,4,4,4]

plt.hist(data,bins=4)
plt.title("HISTOGRAM")
plt.xlabel("value")
plt.ylabel("frequence")
plt.show()
```

```python
#scatter plot
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[5,7,6,8,7]
plt.scatter(x,y)
plt.title("SCATTER")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
#PIE chart
import matplotlib.pyplot as plt

labels=['a','b,','c','d']
sizes=[20,30,25,25]
plt.pie(sizes,labels=labels,autopct='%1.1f%%')
plt.title("PIE CHART")
plt.show()
```

```python
# STACKED BAR CHART
import matplotlib.pyplot as plt

x = ['A', 'B', 'C']
y1 = [3, 5, 2]
y2 = [4, 6, 3]

plt.bar(x, y1, label='Data1')
plt.bar(x, y2, bottom=y1, label='Data2')

plt.title("Stacked Bar Chart")
plt.legend()
plt.show()
```

```python
#AREA PLOT
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,15,25]

plt.fill_between(x,y)
plt.title("Area Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
